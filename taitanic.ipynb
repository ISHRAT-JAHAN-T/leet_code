Machine Learning Use for Prediction 

```python
%pip install scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix , classification_report
from sklearn.tree import plot_tree
```

Determine the features and target variable 

```python
#solit data frame into two part. X contain all features without Survived and y contain only Survived 

X= df.drop(['Survived'], axis=1)
y=df['Survived']
```

```python
print(X)
```

```python
print(y)
```

Split the Dataset to Train & Test set

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
scaler= StandardScaler()
scaler.fit(X_train)
```

Train Model 

```python
# Using Decision tree for making predictions 

model=DecisionTreeClassifier() 
model.fit(X_train, y_train)

```

Predicting Test Data 

```python
y_pred=model.predict(X_test)
```

Evaluating the Model 


```python

confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
model.feature_importances_
```

```python
pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance'])
```

Visualize the tree

```python


plt.figure(figsize=(12, 8), dpi=150)
plot_tree(model, filled=True, feature_names=X.columns.tolist())
```
